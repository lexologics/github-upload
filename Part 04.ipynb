```python
people = {
    'first': ['Corey', 'Jane', 'John'],
    'last': ['Schafer', 'Doe', 'Doe'],
    'email': ['CoreySchafer@gmail.com', 'janedoe@email.com', 'johndoe@email.com']
}
```
